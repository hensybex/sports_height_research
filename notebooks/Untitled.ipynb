```python
import pandas as pd
```

```python
df = pd.read_csv('../data/archive/athlete_events.csv')
```

```python
df['Sport'].value_counts()[:30]
```

```python
dfFootball = df.loc[df['Sport'] == 'Football']
```

```python
dfFootball = dfFootball.loc[pd.isna(dfFootball['Height']) != 1]
```

```python
def medal(string):
    if pd.isna(string):
        return 0
    if string == 'Bronze':
        return 1
    if string == 'Silver':
        return 2
    if string == 'Gold':
        return 3
    else:
        return 100
```

```python
dfFootball['Medal_num'] = dfFootball['Medal'].apply(medal)
```
